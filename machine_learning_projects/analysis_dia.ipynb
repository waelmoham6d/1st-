```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
# from sklearn.linear_model import LinearRegression
from sklearn import svm 
from sklearn.metrics import accuracy_score  
```

```python
corr_mat=dia.corr(method="pearson")
sns.heatmap(corr_mat,annot=True) 
# hight_feature=corr_mat.index[abs(corr_mat['Outcome'])>=0.0].tolist()
# hight_feature.remove("Outcome")
# hight_feature #reduce acc value
```

```python
x=dia.drop(columns='Outcome') 
y=dia['Outcome']
```

```python
scaler=StandardScaler()
scaler.fit(x)
x=scaler.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.0595,stratify=y,random_state=42)
reg=svm.SVC(kernel='poly')
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
